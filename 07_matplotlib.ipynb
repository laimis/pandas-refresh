```python
import pandas as pd
```

```python
births = pd.read_csv('births.csv')
print("before cleaning", births.shape)

quartiles = np.percentile(births['births'], [25, 50, 75])
mu, sig = quartiles[1], 0.74 * (quartiles[2] - quartiles[0])
births = births[(births['births'] > (mu - 5 * sig)) & (births['births'] < (mu + 5 * sig))]
print("after cleaning", births.shape)
```

```python
print("quartiles", quartiles)
print("mu, sig", mu, sig)
```

```python
births.head()
```

```python
births["day"] = births["day"].astype(int)
```

```python
births.index = pd.to_datetime(10000 * births.year
    + 100 * births.month
    + births.day, format='%Y%m%d')

births_by_date = births.pivot_table('births', [births.index.month, births.index.day])
print(births_by_date.shape)
births_by_date.head()
```

```python
births_by_date.index = [pd.datetime(2012, month, day) for (month, day) in births_by_date.index]
print(births_by_date.shape)
births_by_date.head()
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
births_by_date.plot(ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(24, 8))
births_by_date.plot(ax=ax)

style = dict(size=10, color='gray')

keyDatesAndLabels = {
    '2012-01-01': "New Year's Day",
    '2012-07-04': "Independence Day",
    '2012-09-04': "Labor Day",
    '2012-10-31': "Halloween",
    '2012-11-25': "Thanksgiving",
    '2012-12-25': "Christmas"
}

xKeys = []
yKeys = []

for date, label in keyDatesAndLabels.items():
    count = births_by_date.loc[date, "births"].item()
    ax.text(date, count, label, **style)

ax.plot(xKeys, yKeys, 'x', color='red')

ax.set(title='Number of births by day of year', ylabel='average daily births')

ax.xaxis.set_major_locator(mpl.dates.MonthLocator())
ax.xaxis.set_minor_locator(mpl.dates.MonthLocator(bymonthday=15))
ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter('%h'))
```
